```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset and display the first 10 rows of the data
df = pd.read_csv('Spotify_Youtube.csv')
df.head(10)
```

```python
# Check and display the column names
print("List of Column names:")
for col in df.columns:
    print(col)
```

```python
# Display the info of the dataset and make description of the column counts, 
# number of data points, some columns that contain null values, and the datatypes counts.

print("Info of the Data Set:")
df.info()

num_cols = df.shape[1]
print(f"\nNumber of columns: {num_cols}")

num_data_points = df.shape[0]
print(f"Number of data points: {num_data_points}")

null_cols = df.columns[df.isnull().any()].tolist()
print(f"\nColumns with null values: {null_cols}")

dtype_counts = df.dtypes.value_counts().to_dict()
print(f"\nDatatype counts: {dtype_counts}")
```

```python
# Fill NA in likes with 0, NA in comments with 0, and drop others.
df['Likes'] = df['Likes'].fillna(0)
df['Comments'] = df['Comments'].fillna(0)

df1 = df[["Likes", "Comments"]]

df1.head(10)
```

```python
# Describe (descriptive measures: count, mean, std, min, 25%, 50%, 75%, max) the cleaned data.
df.describe()
```

```python
# Explore album type and display counts for album, single, and compilation.
counts = df['Album_type'].value_counts()
print(counts)
```

```python
# Create a pie chart for Album Types as a Title and make some remarks.
counts = df['Album_type'].value_counts()
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%')
plt.title('Album Types')

plt.show()
```
